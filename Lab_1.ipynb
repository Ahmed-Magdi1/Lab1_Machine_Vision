#**1) Import a colored image from your computer (Lab1_1.jpg).**

```python
# Important libraries
import  cv2 as cv  #Define computer vision library
import numpy as np #Define numpy library
from matplotlib import pyplot as plt #Define matplotlib library
```

```python
colored_image = cv.imread('/content/sample_data/Lab_1_1.jpg') #Read the image
rgb_image = cv.cvtColor(colored_image, cv.COLOR_BGR2RGB) #Convert the image from BGR to RGB
plt.imshow(rgb_image) #Show the image
plt.title('Colored_image')  # Set the title
plt.show()  # Show the plot
```

#**2) Find the size of your image.**

```python
# Get the image dimensions (height, width, channels). The number of channels indicates the color depth of the image.
height, width, channels = colored_image.shape

# Print the dimensions of the image
print(f"Width: {width}, Height: {height}, Channels: {channels}")
```

#**3) Resize the image (Lab1_1.jpg) with three different resolutions (100 x 100x 3), (200 x 200 x 3), (400 x 400 x 3).**

```python
# Resize the image to (100 x 100)
image_100x100 = cv.resize(colored_image, (100, 100))
image_200x200 = cv.resize(colored_image, (200, 200))
image_400x400 = cv.resize(colored_image, (400, 400))
```

```python
# Function to display images
def display_image(image, title):
    # Convert from BGR to RGB to display correctly in matplotlib
    rgb_image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
    plt.imshow(rgb_image)  # Display the image
    plt.title(title)  # Set the title
    plt.axis('off')  # Hide axis
    plt.show()  # Show the plot

# Display the resized images
display_image(image_100x100, '100x100 Resolution (3 Channels)')
display_image(image_200x200, '200x200 Resolution (3 Channels)')
display_image(image_400x400, '400x400 Resolution (3 Channels)')
```

#**4) Create a colored image with resolution (320 x 320 x 3) and set all pixels to be black then save it with name (Black.jpg).**

```python
new_colored_image_black = np.ones((320, 320, 3), dtype='uint8')  # Create a new 320x320 image with 3 color channels
plt.imshow(new_colored_image_black) # Display the image
plt.title('320x320 Resolution (3 Channels)')  # The title of the image
plt.axis('off')  # Hide axis
plt.show() # Show the image on the screen
cv.imwrite('Black.jpg', new_colored_image_black) # Save the image with name 'Black.jpg'
```

#**5) Create the same picture as (2) but set all pixels to be white and save it as (White.jpg).**

```python
new_colored_image_white = np.ones((320, 320, 3), dtype='uint8') * 255 # Create a new 320x320 image with 3 color channels
plt.imshow(new_colored_image_white) # Display the image
plt.title('320x320 Resolution (3 Channels)')  # The title of the image
plt.axis('off')  # Hide axis
plt.show() # Show the image on the screen
cv.imwrite('White.jpg', new_colored_image_white) # Save the image with name 'Black.jpg'
```

#**6) Using the image White.jpg, do the following:**

**- Write the text "BUE" in the center of the image using black font. The font size should be approximately two-thirds of the image height.**

```python
# Calculate font scale as two-thirds of the image height
height, width = new_colored_image_white.shape[:2]
font_scale = (2 / 3) * (height/100)  # 2/3 of image height
font_thickness = 3  # Adjust thickness for clarity
font = cv.FONT_HERSHEY_COMPLEX  # Font type

# Text to write
text = "BUE"

# Calculate text size and position
text_size = cv.getTextSize(text, font, font_scale, font_thickness)[0]
text_x = (width - text_size[0]) // 2  # Centered horizontally
text_y = (height + text_size[1]) // 2  # Centered vertically

# Put the "BUE" text in the center of the image in black font
BUE_text = cv.putText(new_colored_image_white, text, (text_x, text_y), font, font_scale, (0, 0, 0), font_thickness)

# Display the image with the text
plt.imshow(new_colored_image_white)
plt.title('320x320 Resolution (3 Channels) with "BUE"')  # The title of the image
plt.axis('off')  # Hide axis
plt.show()  # Show the image
```

**- Draw a black rectangle around the "BUE" text.**

```python
# Top-left corner and bottom-right corner coordinates for the rectangle
top_left = (text_x - 10, text_y - text_size[1] - 10)  # Padding of 10 pixels around the text
bottom_right = (text_x + text_size[0] + 10, text_y + 10)  # Padding of 10 pixels around the text

Rectangle_image = cv.rectangle(BUE_text, top_left, bottom_right, (0, 0, 0), thickness=2)  # Black rectangle

plt.imshow(Rectangle_image)  # Display the image with the text and rectangle
plt.title('320x320 Resolution (3 Channels) with "BUE" and Rectangle')  # The title of the image
plt.axis('off')  # Hide axis
plt.show()  # Show the image
```

**- Draw black circles near the borders of the image.**

```python
circle_radius = 10  # Radius of the circles
# Coordinates for the circles (corners and center of each edge)
circle_positions = [
    (circle_radius, circle_radius),  # Top-left
    (width - circle_radius, circle_radius),  # Top-right
    (circle_radius, height - circle_radius),  # Bottom-left
    (width - circle_radius, height - circle_radius),  # Bottom-right
    (width // 2, circle_radius),  # Top-center
    (width // 2, height - circle_radius),  # Bottom-center
    (circle_radius, height // 2),  # Left-center
    (width - circle_radius, height // 2)  # Right-center
]

# Draw each circle using for loop
for pos in circle_positions:
    circle_image = cv.circle(Rectangle_image, pos, circle_radius, (0, 0, 0), thickness=-1)  # Filled circle


plt.imshow(circle_image)  # Display the image with the text, rectangle, and circles
plt.title('320x320 Resolution (3 Channels) with "BUE", Rectangle, and Circles')  # The title of the image
plt.axis('off')  # Hide axis
plt.show()  # Show the image
```

**-Write your name in blue anywhere on the image.**

```python
name = "AHMED MAGDI"
name_font_scale = 0.5  # Scale for the name font size
name_thickness = 2  # Thickness for the name font

# Calculate text size and position for the name
name_size = cv.getTextSize(name, font, name_font_scale, name_thickness)[0]  # Get the size of the text
name_x = 100  # Position from the left edge
name_y = height - 50  # Position from the bottom edge

# Put the name in blue on the image
myname_image = cv.putText(circle_image, name, (name_x, name_y), font, name_font_scale, (0, 0, 255), name_thickness)  # Blue text

plt.imshow(myname_image)  # Display the image with the text, rectangle, circles, and name
plt.axis('off')  # Hide axis
plt.show()
```

**- Draw an ellipse around your name.**

```python
center_coordinates = (name_x + name_size[0] // 2, name_y - name_size[1] // 2)  # Center of the ellipse
axes_lengths = (name_size[0] // 2 + 10, name_size[1] // 2 + 10)  # Semi-major and semi-minor axes
angle = 0  # Angle of rotation
start_angle = 0  # Starting angle
end_angle = 360  # Ending angle
ellipse_image = cv.ellipse(myname_image, center_coordinates, axes_lengths, angle, start_angle, end_angle, (0, 0, 255), thickness=2)  # Draw the ellipse in blue

plt.imshow(ellipse_image)  # Display the image
plt.axis('off')  # Hide axis
plt.show()  # Show the image
```

**- Save the modified image with your name as (yourname.jpg).**

```python
Final_image = cv.cvtColor(ellipse_image, cv.COLOR_RGB2BGR)  # Convert back to BGR for saving

cv.imwrite('AHMED_MAGDI.jpg', Final_image)  # Save the image with the name "AHMED_MAGDI.jpg"
```

# **7) Do the following operation to (,) and explain the effect of each operation.**

```python
# Re-load the final image
named_image = cv.imread('/content/AHMED_MAGDI.jpg')  # Load the image
Function_image = cv.cvtColor(named_image, cv.COLOR_BGR2RGB) #Convert the image from BGR to RGB
```

# **a) 2(,)**

```python
multiplied_image = 2 * Function_image  # Multiply each pixel by 2

# Ensure pixel values don't exceed the maximum (255 for 8-bit image)
multiplied_image = np.clip(multiplied_image, 0, 255).astype(np.uint8)

plt.imshow(multiplied_image)  # Display the image
plt.title('2 * f(u,v)')  # The title of the image
plt.axis('off')  # Hide axis
plt.show() # Show the image
```

# Explanation

By multiplying the function (,) by two, the pixel values in the image are doubled, resulting in increased brightness. However, in my case, no visible change will occur because the image consists of only three colors: black, white, and blue. In the case of the :

Black areas: No change will occur because black has zero intensity in all channels (RGB). When multiplied by 2, it will remain zero, meaning the black areas will stay black.

White areas: Similarly, there will be no change in the white areas. White has an intensity of 255 in all three channels (RGB), and multiplying this by 2 would result in 510. Since the maximum value for an 8-bit image is 255, the pixel values will remain at 255, so the white areas will remain white at maximum brightness.

Blue areas: Blue areas typically have a high intensity in the blue channel while the red and green channels are at zero. In my case, where the blue areas are set to (0, 0, 255) in RGB, multiplying the blue channel by 2 would result in 510. However, since 255 is the maximum, the blue areas will also remain unchanged, at their maximum intensity of 255 in the blue channel.

The multiplication by 2 affects mid-range pixel values, not those at the maximum or zero values.

# b) (,) / 2

```python
divided_image = Function_image / 2  # Divide each pixel by 2

# Ensure the pixel values stay within the valid range for uint8 (0 to 255)
divided_image = np.clip(divided_image, 0, 255).astype(np.uint8)

plt.imshow(divided_image)  # Display the image
plt.title('f(u,v) / 2')  # The title of the image
plt.axis('off')  # Hide axis
plt.show()  # Show the image
```

# Explanation

The image will generally appear darker because dividing the pixel intensities by 2 reduces their brightness. However, the black areas will remain unchanged since they already have a pixel intensity of zero. Dividing zero by 2 still results in zero, so the black areas will stay black.

# c) (,)+100

```python
adding = cv.add(Function_image, 100)  # Add 100 to each pixel
plt.imshow(adding)  # Display the image
plt.title('f(u,v) + 100')  # The title of the image
plt.axis('off')  # Hide axis
plt.show()  # Show the image
```

# Explanation

The entire image will become brighter, especially in the darker areas. Black elements will turn into light gray, and the blue text will become lighter blue.

# d) (,)-100

```python
sub = cv.subtract(Function_image, 100)  # Subtract 100 from each pixel
plt.imshow(sub)  # Display the image
plt.title('f(u,v) - 100')  # The title of the image
plt.axis('off')  # Hide axis
plt.show()  # Show the image
```

# Explanation
By subtracting 100 from each pixel, the image will darken. White areas will turn into light gray since the values will decrease from 255 to 155. Black areas will stay black as they have zero value. Blue areas like my name will darken as their pixel values drop, turning into dark blue.

# e) 255  (,)

```python
sub = cv.subtract(255, Function_image) # Subtract (255  (,))
plt.imshow(sub)  # Display the image
plt.title('255 - f(u,v)')  # The title of the image
plt.axis('off')  # Hide axis
plt.show()  # Show the image
```

# Explanation

White will turn to black as (255 - 255 = 0), which represents (0, 0, 0) RGB,
black to white as (255 - 0 = 255), which represents (255, 255, 255) RGB, and for the blue, it will turn into yellow as (255 - 0, 255 - 0, 255 - 255), which represents (255, 255, 0) RGB.

# f) 0.3+0.6+0.1

```python
B, G, R = cv.split(Function_image) # Split the image into R, G, and B channels

grayscale_image = 0.3 * R + 0.6 * G + 0.1 * B  # Calculate the weighted sum of the color channels

grayscale_image = np.clip(grayscale_image, 0, 255).astype(np.uint8)  # Ensure pixel values stay within the range of (0 - 255)

plt.imshow(grayscale_image, cmap='gray')  # Display the color map in grayscale
plt.title('0.3 * fR + 0.6 * fG + 0.1 * fB')  # The title of the image
plt.axis('off')  # Hide axis
plt.show()  # Show the image
```

# Explanation

By Applying the formula (0.3 * R + 0.6 * G + 0.1 * B) to calculate the weighted sum of the color channels in grayscale:


For Black areas: since black is (0, 0, 0) in RGB. The formula 0.3 x 0 + 0.6 x 0 + 0.1 x 0 = 0, so black areas remain black.


For White areas: since white is (255, 255, 255) in RGB. The formula 0.3 x 255 + 0.6 x 255 + 0.1 x 255 = 255, so White areas remain white.


For Blue areas: since blue is (0, 0, 255) in RGB. The formula 0.3 x 0 + 0.6 x 0 + 0.1 x 255 = 25.5, which will be equal to 25 in 8-bit int, so blue areas will become dark gray.
